```python
h, w = image_3.shape
boxes = pytesseract.image_to_boxes(image_3) 
for b in boxes.splitlines():
    b = b.split(' ')
    image_3 = cv2.rectangle(image_3, (int(b[1]), h - int(b[2])), (int(b[3]), h - int(b[4])), 0, 2)

image_3_pil = Image.fromarray(image_3)
display(image_3_pil)
```
