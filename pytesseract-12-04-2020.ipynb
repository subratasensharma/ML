```python
# import the necessary packages
from PIL import Image
import pytesseract
import cv2
import numpy as np
```

```python
image_1 = cv2.imread("INVOICE2.jpg") ## Loading image

print(image_1.shape)
print(image_1[100,100,0])

total_pixels = image_1.shape[0]*image_1.shape[1]*image_1.shape[2]
total_pixels_1 = image_1.size
min_value_pixel = image_1.min()
max_value_pixel = image_1.max()
print('total no. of pixels = ', total_pixels)
print('total no. of pixels thru .size operator = ', total_pixels_1)
print('minimum value of pixel = ', min_value_pixel)
print('maximum value of pixel = ', max_value_pixel)
np.count_nonzero(image_1) # counting non zero pixels
```

```python
image_1_gray = cv2.cvtColor(image_1, cv2.COLOR_BGR2GRAY) ## Convert image to b_w
print(image_1_gray.shape)
print(image_1_gray[100,100])

total_pixels_gray = image_1_gray.shape[0]*image_1_gray.shape[1]
min_value_pixel_gray = image_1_gray.min()
max_value_pixel_gray = image_1_gray.max()
print('total no. of gray pixels = ', total_pixels_gray)
print('minimum value of gray pixel = ', min_value_pixel_gray)
print('maximum value of gray pixel = ', max_value_pixel_gray)
np.count_nonzero(image_1_gray) # counting non zero pixels
```

```python
img_1_gray = Image.fromarray(image_1_gray)
#display(img_1_gray)
type(img_1_gray)
```

```python
img_1_gray_sharp = img_1_gray.point(lambda x: 0 if x < 200 else 255, '1')
display(img_1_gray_sharp)
```

```python
image_1_gray_crop_1 = image_1_gray[50:301, 100:]
print(image_1_gray_crop_1.min())
print(image_1_gray_crop_1.max())
print(image_1_gray_crop_1.shape)
```

```python
img_1_gray_crop_1 = Image.fromarray(image_1_gray_crop_1)
img_1_gray_crop_1_sharp = img_1_gray_crop_1.point(lambda x: 0 if x < 200 else 255, '1')
display(img_1_gray_crop_1_sharp)
```

```python
from pytesseract import image_to_string

text_1 = pytesseract.image_to_string(img_1_gray_crop_1)

text_1_sharp = pytesseract.image_to_string(img_1_gray_crop_1_sharp)

print(text_1)

print('==================================')

print(text_1_sharp)
```
