```python

image_1 = cv2.imread("/home/scar3crow/Downloads/INVOICES/1.jpg", 0) ## Loading image directly as black & white.

(h, w) = image_1.shape # h = no. of rows, w = no. of columns

# find texts upto 5% of height

h_find = int(h/4)

## Converting image to text and create pandas dataframe with all text cordinates and making texts in lower case

image_1_data = pytesseract.image_to_data(image_1[0:h_find, :], output_type=Output.DICT)
image_1_pd = pd.DataFrame(image_1_data)

image_1_pd_sorted =image_1_pd.sort_values(['top', 'left']).reset_index()
image_1_pd_sorted['text'] = image_1_pd_sorted['text'].map(lambda x: x if type(x)!=str else x.lower()) ## make Items in lower case

image_1_pd_sorted['text'].replace('', np.nan, inplace = True)
image_1_pd_sorted['text'].replace(' ', np.nan, inplace = True)
image_1_pd_sorted['text'].replace('  ', np.nan, inplace = True)
image_1_pd_sorted.dropna(subset=['text'], inplace = True)
image_1_pd_sorted = image_1_pd_sorted.reset_index(drop=True)
image_1_pd_sorted

#image_pil_1 = Image.fromarray(image_1[0:h_find, :])

#display(image_pil_1) 
```
